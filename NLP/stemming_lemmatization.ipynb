```python
sentence = """amaze amazed amazing"""
```

```python
words_to_stem = sentence.split()
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
ps = PorterStemmer()
ls = LancasterStemmer()
```

```python
stemmed = [ls.stem(word) for word in words_to_stem]

stemmed
```

```python
import nltk

nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
([lemmatizer.lemmatize(word) for word in words_to_stem])
```

```python
from nltk.corpus import wordnet
```

```python
([lemmatizer.lemmatize(word, wordnet.VERB) for word in words_to_stem])
```

## Handle missing data

### Subtask:
Address the missing data identified earlier by either dropping columns with a high percentage of missing values or imputing missing values for relevant columns.


**Reasoning**:
Drop columns with over 50% missing values and impute missing values for selected relevant columns. Then, verify the missing values.



```python
# Drop columns with over 50% missing values
missing_threshold = 50
cols_to_drop = missing_df[missing_df["Percentage"] > missing_threshold].index.tolist()
df_cleaned = df.drop(columns=cols_to_drop)

# Impute missing values for relevant columns with smaller percentage of missing data
# Numerical columns: impute with median
numerical_cols_to_impute = ['release_clause_eur']
for col in numerical_cols_to_impute:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

# Categorical and Date columns: impute with mode or a placeholder
categorical_date_cols_to_impute = ['club_joined_date', 'league_level', 'club_team_id', 'club_name', 'club_position', 'club_jersey_number', 'club_contract_valid_until_year', 'league_name', 'league_id']
for col in categorical_date_cols_to_impute:
    if col in df_cleaned.columns:
        # For numerical IDs, mode is appropriate. For names/dates, mode or a placeholder like 'Unknown'/'Missing'
        if df_cleaned[col].dtype in ['int64', 'float64']:
             df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])
        else:
             df_cleaned[col] = df_cleaned[col].fillna('Missing')


# Verify missing values
missing_cleaned = df_cleaned.isnull().sum()
missing_percent_cleaned = (missing_cleaned / len(df_cleaned)) * 100
missing_df_cleaned = pd.DataFrame({"Missing Count": missing_cleaned, "Percentage": missing_percent_cleaned}).sort_values(by="Percentage", ascending=False)

print("\n=== Missing Data After Cleaning ===")
display(missing_df_cleaned[missing_df_cleaned["Missing Count"] > 0])

df = df_cleaned # Update the original DataFrame reference
```

**Reasoning**:
Impute the remaining relevant columns with missing values using appropriate strategies and then verify that missing values have been handled.



```python
# Impute missing values for the remaining relevant columns (pace, shooting, passing, dribbling, defending, physic)
# These are numerical attributes, so imputation with the median is a reasonable strategy.
attributes_to_impute = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
for col in attributes_to_impute:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

# Verify that the missing values have been handled
missing_cleaned_final = df_cleaned.isnull().sum()
missing_percent_cleaned_final = (missing_cleaned_final / len(df_cleaned)) * 100
missing_df_cleaned_final = pd.DataFrame({"Missing Count": missing_cleaned_final, "Percentage": missing_percent_cleaned_final}).sort_values(by="Percentage", ascending=False)

print("\n=== Missing Data After Final Cleaning ===")
display(missing_df_cleaned_final[missing_df_cleaned_final["Missing Count"] > 0])

df = df_cleaned # Update the original DataFrame reference
```

## Summary:

### Data Analysis Key Findings

* **Player Attributes by Position:** Our box plots clearly showed that key attributes like pace, shooting, passing, dribbling, defending, and physic are not uniformly distributed. For example, players in attacking positions (like ST, CF, LW, RW) generally had higher pace, shooting, and dribbling ratings, while defensive players (like CB, LCB, RCB) had higher defending and sometimes physic ratings. This visual evidence directly supports the expected roles of these positions.

* **Player Value and Wage Determinants:** The scatter plots revealed strong positive correlations between 'overall' rating and both 'value_eur' and 'wage_eur'. Similarly, 'potential' also showed a significant positive correlation with value and wage. The correlation matrix further quantified these relationships. Age had a less linear relationship; scatter plots showed a general increase in value and wage with age up to a certain point, then potentially a decrease. Box plots comparing 'value_eur' and 'wage_eur' by 'club_position' indicated that attacking and central midfield positions tend to have higher median values and wages.

* **Team and League Statistics:** By grouping the data and calculating averages and sums, we saw clear patterns. The bar plots of the top 10 clubs by average overall rating and total squad value demonstrated the dominance of well-known teams like Real Madrid, Paris Saint-Germain, and Arsenal. Similarly, the bar plots for leagues highlighted that leagues like the Premier League, La Liga, and Serie A have higher average overall ratings and total values, reflecting the concentration of high-quality players and financial power in these leagues.

* **Player Growth and Potential:** The scatter plot of 'overall' vs. 'age' showed a general trend of increasing overall rating with age, often leveling off or slightly decreasing in later years. The scatter plot of 'potential' vs. 'overall' illustrated that players with higher potential tend to have higher overall ratings, but there's also a spread, indicating that some players with high potential haven't yet reached their peak overall rating.

* **Player Preferred Foot and Body Type:** The count plots showed the distribution of players by 'preferred_foot' (mostly Right) and 'body_type'. While the average 'overall' rating difference between preferred feet was small, the box plots of attributes like 'power_strength', 'movement_agility', 'movement_balance', 'movement_acceleration', and 'movement_sprint_speed' by 'body_type' indicated that certain body types are associated with different distributions of these physical attributes (e.g., 'Stocky' types tending to have higher strength, 'Lean' types higher agility/acceleration).

* **Handling Missing Data:** We successfully addressed missing data by dropping columns with a high percentage of missing values and imputing the remaining relevant numerical columns with their median. This ensured that our analysis was performed on a dataset without missing values in the columns used for analysis.

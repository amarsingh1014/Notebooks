```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
X=data.drop(columns=["Patient Number","Expert Diagnose"])
y=data["Expert Diagnose"]
X_encoded=X.apply(LabelEncoder().fit_transform)
y_encoded=LabelEncoder().fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y_encoded,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=2500,random_state=42,class_weight='balanced')
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred,average='weighted'))
print("Precision:",precision_score(y_test,y_pred,average='weighted'))
print("F1:",f1_score(y_test,y_pred,average='weighted'))
print("\nConfusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("\nClassification Report:\n",classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_estimators':[1500,2000,2500],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['sqrt','log2'],
    'bootstrap':[True,False]
}
rf=RandomForestClassifier(random_state=42,class_weight='balanced')
grid_search=GridSearchCV(estimator=rf,param_grid=param_grid,cv=5,scoring='f1_weighted',verbose=0,n_jobs=-1)
grid_search.fit(X_train,y_train)
print("Best Parameters are:",grid_search.best_params_)
```

```python
best_model=RandomForestClassifier(**grid_search.best_params_,random_state=42,class_weight='balanced')
best_model.fit(X_train,y_train)
```

```python
y_pred_best=best_model.predict(X_test)
print("Tuned Accuracy:",accuracy_score(y_test,y_pred_best))
print("Tuned F1 Score:",f1_score(y_test,y_pred_best,average='weighted'))
print("Tuned Recall Score:",recall_score(y_test,y_pred_best,average='weighted'))
print("Tuned Precision Score:",precision_score(y_test,y_pred_best,average='weighted'))
print("\nTuned Confusion Matrix:\n",confusion_matrix(y_test,y_pred_best))
print("\nTuned Classification Report:\n",classification_report(y_test,y_pred_best))
```
